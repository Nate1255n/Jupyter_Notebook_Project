## Cleaning up the Data

#### We only want USA Races, 50km or 50MI, for the year 2020

```python
# Filtering the data for 50km
df.loc[df['Event distance/length'] == '50km']
```

```python
# Combine 50km/50mi with isin
```

```python
# filter for 50km/50mi

df.loc[df['Event distance/length'].isin(['50km', '50mi'])] 
```

```python
# filtering for 50km or 50mi and the year 2020

df.loc[(df['Event distance/length'].isin(['50km', '50mi'])) &
        (df['Year of event']== 2020)]
```

```python
# Take the brackets out of "(USA)" out so we can filter by it

df.loc[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)'] \
['Event name'].str.split('(').str.get(1) \
.str.split(')').str.get(0)
```

```python
df.loc[df['Event name'].str.split('(').str.get(1) \
        .str.split(')').str.get(0) == 'USA']
```

```python
# Combine all the filters together 
```

```python
df.loc[(df['Event distance/length'].isin(['50km', '50mi'])) &
        (df['Year of event']== 2020) & 
        (df['Event name'].str.split('(').str.get(1) \
        .str.split(')').str.get(0) == 'USA')]
```

```python
# put the filter under a new df, df2, and look at the first five rows
df2 = df.loc[(df['Event distance/length'].isin(['50km', '50mi'])) &
        (df['Year of event']== 2020) & 
        (df['Event name'].str.split('(').str.get(1) \
        .str.split(')').str.get(0) == 'USA')]

df2.head(5)
```

```python
# Lets see if the rows and col changed so we know we did it right
df2.shape
```

```python
# completely remove the "(USA)" from event name and then get the first index
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(5)
```

```python
# Clean up the athlete age 
```

```python
# get the age of the athletes under a new column 
df2['Athlete_Age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove the "h" from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# drop unnecssary cols: Athlete Club, Althlete Country, Athlete Year of birth, Athlete Age Category
# Axis = 1 : so it knows we're dropping the cols
```

```python
df2 = df2.drop(['Athlete club','Athlete country', 'Athlete year of birth', 
       'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
# clean up null values
# isna() : shows all the null values 
# .sum() : will total up all the null values
```

```python
df2.isna().sum()
```

```python
# Shows the null values in the 'Athlete_Age' col
df2.loc[df2['Athlete_Age'].isna()==1]
```

```python
# drop the null values and overwrite the original df2
df2 = df2.dropna()
```

## Extra Analysis 

```python
# Difference in speed for the 50k & 50mi by male and female
```

```python
df3.groupby(['race_length', 'athlete_gender']) \
            .agg({'athlete_avg_speed':'mean'}) \
            .reset_index()
```

```python
# What age groups are the best in the 50m Race (20+ races min) (show 15)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age') \
            ['athlete_avg_speed'].agg(['mean', 'count']) \
            .sort_values('mean', ascending=False).query('count > 19') \
            .head(15)
```

```python
# What age groups are the worst in the 50m Race (20 + races min) (show 20)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age') \
            ['athlete_avg_speed'].agg(['mean', 'count']) \
            .sort_values('mean', ascending=True).query('count > 19') \
            .head(20)
```

```python
# Seasons for the data ->  Are runners slower in the Summer than Winter?

#Spring 3-5
#Summer 6-8
#Fall 9-11
#Winter 12-2

# Spilt between two decimals
```

```python
# create a new col called race_month by extract the month number from race_day
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
# Create a new col called 'race_season' and name them by the season (ex: Fall, Summer, etc.) 
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' \
                                                       if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
# we find out runners are slower in the Summer than Winter but we notice there are more racers in Winter than Summer
df3.groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

```python
# We try to see if it's true for '50 miler only' and the same seems to be the true
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
